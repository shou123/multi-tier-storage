# I/O Trace Data Analysis: Data Size vs Latency

This notebook analyzes the I/O trace data from `wdev_3.revised` to examine the relationship between operation type (WS/RS), data size, and latency.

**Trace File Columns:**
- timestamp: Time when the I/O request was issued
- operation_type: WS (write) or RS (read)
- LBA: Logical Block Address
- data_size: Size of data in bytes
- access_pattern: seq (sequential) or rand (random)
- inter_arrival: Time since last I/O
- latency: Response time of the I/O request
- idle: Idle time

**Analysis Goals:**
1. Parse and load trace data into a structured format
2. Extract WS/RS operations with data size and latency
3. Perform exploratory data analysis (EDA)
4. Visualize data size vs latency relationships
5. Analyze I/O patterns by operation type and access pattern
6. Provide statistical summary and correlations

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings

warnings.filterwarnings('ignore')

# Set style for better visualizations
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10
```

## Section 1: Load and Parse Trace Data

Read the trace file and parse all columns into a structured DataFrame.

```python
# Load and parse the trace data
trace_file = '/home/houjason/project/multi-tier-storage-systems-simulator/wdev_3.revised'

# Define column names based on the trace format
columns = ['timestamp', 'operation_type', 'LBA', 'data_size', 'access_pattern', 
           'inter_arrival', 'latency', 'idle_time']

# Read the trace file with space delimiter
df = pd.read_csv(trace_file, delimiter='\s+', names=columns, dtype={
    'timestamp': float,
    'operation_type': str,
    'LBA': int,
    'data_size': int,
    'access_pattern': str,
    'inter_arrival': float,
    'latency': float,
    'idle_time': float
})

print("Trace Data Loaded Successfully!")
print(f"Total records: {len(df)}")
print(f"\nDataFrame shape: {df.shape}")
print(f"\nFirst 10 rows:")
print(df.head(10))
print(f"\nData types:")
print(df.dtypes)
print(f"\nBasic statistics:")
print(df.describe())
```

## Section 2: Extract Relevant Columns

Extract only WS (write) and RS (read) operations with their corresponding data_size and latency values. Filter out any incomplete or malformed records.

```python
# Filter for WS and RS operations only
io_operations = df[df['operation_type'].isin(['WS', 'RS'])].copy()

print(f"Total I/O operations (WS + RS): {len(io_operations)}")
print(f"\nOperation type distribution:")
print(io_operations['operation_type'].value_counts())

# Check for missing values
print(f"\nMissing values:")
print(io_operations.isnull().sum())

# Remove any rows with NaN values in critical columns
io_operations = io_operations.dropna(subset=['operation_type', 'data_size', 'latency'])

print(f"\nAfter removing incomplete records: {len(io_operations)} records")

# Convert latency from seconds to milliseconds for better readability
io_operations['latency_ms'] = io_operations['latency'] * 1000

# Create data size categories (in bytes)
def categorize_data_size(size):
    if size <= 8:
        return '8B'
    elif size <= 16:
        return '16B'
    elif size <= 128:
        return '128B'
    else:
        return 'Other'

io_operations['data_size_category'] = io_operations['data_size'].apply(categorize_data_size)

print(f"\nData size distribution:")
print(io_operations['data_size'].value_counts().sort_index())
print(f"\nData size categories:")
print(io_operations['data_size_category'].value_counts())

print(f"\nAccess pattern distribution:")
print(io_operations['access_pattern'].value_counts())

print(f"\nExtracted data sample:")
print(io_operations[['timestamp', 'operation_type', 'data_size', 'access_pattern', 'latency_ms']].head(10))
```

## Section 3: Exploratory Data Analysis (EDA)

Calculate descriptive statistics for latency grouped by operation type and data size categories.

```python
# Overall latency statistics
print("=" * 80)
print("OVERALL LATENCY STATISTICS")
print("=" * 80)
print(f"Count: {io_operations['latency_ms'].count()}")
print(f"Mean:   {io_operations['latency_ms'].mean():.4f} ms")
print(f"Median: {io_operations['latency_ms'].median():.4f} ms")
print(f"Min:    {io_operations['latency_ms'].min():.4f} ms")
print(f"Max:    {io_operations['latency_ms'].max():.4f} ms")
print(f"Std:    {io_operations['latency_ms'].std():.4f} ms")
print(f"P95:    {io_operations['latency_ms'].quantile(0.95):.4f} ms")
print(f"P99:    {io_operations['latency_ms'].quantile(0.99):.4f} ms")

# Latency statistics grouped by operation type
print("\n" + "=" * 80)
print("LATENCY STATISTICS BY OPERATION TYPE")
print("=" * 80)
latency_by_op = io_operations.groupby('operation_type')['latency_ms'].describe()
print(latency_by_op)

# Latency statistics grouped by data size
print("\n" + "=" * 80)
print("LATENCY STATISTICS BY DATA SIZE CATEGORY")
print("=" * 80)
latency_by_size = io_operations.groupby('data_size_category')['latency_ms'].describe()
print(latency_by_size)

# Latency statistics grouped by operation type and data size
print("\n" + "=" * 80)
print("LATENCY STATISTICS BY OPERATION TYPE AND DATA SIZE")
print("=" * 80)
latency_by_op_size = io_operations.groupby(['operation_type', 'data_size_category'])['latency_ms'].describe()
print(latency_by_op_size)

# Latency statistics grouped by access pattern
print("\n" + "=" * 80)
print("LATENCY STATISTICS BY ACCESS PATTERN")
print("=" * 80)
latency_by_pattern = io_operations.groupby('access_pattern')['latency_ms'].describe()
print(latency_by_pattern)
```

## Section 4: Visualize Data Size vs Latency

Create scatter plots and line charts showing the relationship between data size and latency for both WS and RS operations.

```python
# Figure 1: Scatter plot - Data Size vs Latency (colored by operation type)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: All data points
ax1 = axes[0]
for op_type in ['WS', 'RS']:
    data = io_operations[io_operations['operation_type'] == op_type]
    ax1.scatter(data['data_size'], data['latency_ms'], alpha=0.5, s=30, label=op_type)

ax1.set_xlabel('Data Size (bytes)', fontsize=11, fontweight='bold')
ax1.set_ylabel('Latency (ms)', fontsize=11, fontweight='bold')
ax1.set_title('Data Size vs Latency by Operation Type', fontsize=12, fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Log scale for better visualization
ax2 = axes[1]
for op_type in ['WS', 'RS']:
    data = io_operations[io_operations['operation_type'] == op_type]
    ax2.scatter(data['data_size'], data['latency_ms'], alpha=0.5, s=30, label=op_type)

ax2.set_xlabel('Data Size (bytes)', fontsize=11, fontweight='bold')
ax2.set_ylabel('Latency (ms)', fontsize=11, fontweight='bold')
ax2.set_title('Data Size vs Latency (Log Scale)', fontsize=12, fontweight='bold')
ax2.set_yscale('log')
ax2.legend()
ax2.grid(True, alpha=0.3, which='both')

plt.tight_layout()
plt.savefig('/home/houjason/project/multi-tier-storage-systems-simulator/01_data_size_vs_latency.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure 1 saved: 01_data_size_vs_latency.png")
```

```python
# Figure 2: Line plot - Average latency by data size
fig, ax = plt.subplots(figsize=(14, 6))

# Calculate average latency for each data size and operation type
avg_latency = io_operations.groupby(['data_size', 'operation_type'])['latency_ms'].mean().reset_index()

for op_type in ['WS', 'RS']:
    data = avg_latency[avg_latency['operation_type'] == op_type]
    data_sorted = data.sort_values('data_size')
    ax.plot(data_sorted['data_size'], data_sorted['latency_ms'], marker='o', linewidth=2, 
            markersize=8, label=op_type)

ax.set_xlabel('Data Size (bytes)', fontsize=11, fontweight='bold')
ax.set_ylabel('Average Latency (ms)', fontsize=11, fontweight='bold')
ax.set_title('Average Latency by Data Size and Operation Type', fontsize=12, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('/home/houjason/project/multi-tier-storage-systems-simulator/02_avg_latency_by_size.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure 2 saved: 02_avg_latency_by_size.png")
```

```python
# Figure 3: Box plot by data size category and operation type
fig, ax = plt.subplots(figsize=(14, 6))

# Prepare data for box plot
data_for_box = io_operations[io_operations['latency_ms'] < io_operations['latency_ms'].quantile(0.95)]

sns.boxplot(data=data_for_box, x='data_size_category', y='latency_ms', hue='operation_type', ax=ax)

ax.set_xlabel('Data Size Category', fontsize=11, fontweight='bold')
ax.set_ylabel('Latency (ms)', fontsize=11, fontweight='bold')
ax.set_title('Latency Distribution by Data Size Category and Operation Type (95th percentile)', 
             fontsize=12, fontweight='bold')
ax.legend(title='Operation Type', fontsize=10)
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('/home/houjason/project/multi-tier-storage-systems-simulator/03_boxplot_latency_by_category.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure 3 saved: 03_boxplot_latency_by_category.png")
```

## Section 5: Analyze I/O Patterns by Type

Compare latency distributions between sequential and random access patterns. Create box plots and histograms to show latency variations.

```python
# Figure 4: Histograms - Latency distribution by operation type and access pattern
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for idx, op_type in enumerate(['WS', 'RS']):
    for jdx, pattern in enumerate(['seq', 'rand']):
        data = io_operations[(io_operations['operation_type'] == op_type) & 
                            (io_operations['access_pattern'] == pattern)]['latency_ms']
        
        ax = axes[idx, jdx]
        
        # Use log scale for latency to better see distribution
        ax.hist(np.log10(data + 0.0001), bins=50, edgecolor='black', alpha=0.7, color='steelblue')
        ax.set_xlabel('Latency (log10 ms)', fontsize=10, fontweight='bold')
        ax.set_ylabel('Frequency', fontsize=10, fontweight='bold')
        ax.set_title(f'{op_type} - {pattern.upper()} Access\n(n={len(data)}, mean={data.mean():.4f}ms)', 
                    fontsize=11, fontweight='bold')
        ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('/home/houjason/project/multi-tier-storage-systems-simulator/04_latency_histograms.png', dpi=300, bbox_inches='tight')
plt.show()

print("Figure 4 saved: 04_latency_histograms.png")
```

## Section 6: Statistical Summary

Generate correlation analysis between data size and latency, and provide summary tables showing average latency for different data size ranges and operation types.

```python
# Correlation analysis between data size and latency
print("=" * 80)
print("CORRELATION ANALYSIS: Data Size vs Latency")
print("=" * 80)

# Overall correlation
overall_corr = io_operations['data_size'].corr(io_operations['latency_ms'])
print(f"\nOverall Correlation (Pearson): {overall_corr:.6f}")

# Correlation by operation type
print("\nCorrelation by Operation Type:")
for op_type in ['WS', 'RS']:
    op_data = io_operations[io_operations['operation_type'] == op_type]
    corr = op_data['data_size'].corr(op_data['latency_ms'])
    print(f"  {op_type}: {corr:.6f}")

# Correlation by access pattern
print("\nCorrelation by Access Pattern:")
for pattern in ['seq', 'rand']:
    pattern_data = io_operations[io_operations['access_pattern'] == pattern]
    corr = pattern_data['data_size'].corr(pattern_data['latency_ms'])
    print(f"  {pattern.upper()}: {corr:.6f}")

# Statistical test: t-test between WS and RS latencies
print("\n" + "=" * 80)
print("STATISTICAL TESTS")
print("=" * 80)

ws_latency = io_operations[io_operations['operation_type'] == 'WS']['latency_ms']
rs_latency = io_operations[io_operations['operation_type'] == 'RS']['latency_ms']

t_stat, p_value = stats.ttest_ind(ws_latency, rs_latency)
print(f"\nT-Test (WS vs RS Latency):")
print(f"  t-statistic: {t_stat:.6f}")
print(f"  p-value: {p_value:.6e}")
print(f"  Result: {'Significantly different' if p_value < 0.05 else 'Not significantly different'}")

# t-test between seq and rand
seq_latency = io_operations[io_operations['access_pattern'] == 'seq']['latency_ms']
rand_latency = io_operations[io_operations['access_pattern'] == 'rand']['latency_ms']

t_stat2, p_value2 = stats.ttest_ind(seq_latency, rand_latency)
print(f"\nT-Test (SEQ vs RAND Latency):")
print(f"  t-statistic: {t_stat2:.6f}")
print(f"  p-value: {p_value2:.6e}")
print(f"  Result: {'Significantly different' if p_value2 < 0.05 else 'Not significantly different'}")
```

```python
# Summary table 1: Average latency by data size range and operation type
print("\n" + "=" * 80)
print("SUMMARY TABLE 1: Average Latency by Data Size and Operation Type")
print("=" * 80)

summary_table1 = io_operations.groupby(['data_size_category', 'operation_type']).agg({
    'latency_ms': ['count', 'mean', 'median', 'min', 'max', 'std']
}).round(4)

print(summary_table1)

# Summary table 2: Average latency by data size and access pattern
print("\n" + "=" * 80)
print("SUMMARY TABLE 2: Average Latency by Data Size and Access Pattern")
print("=" * 80)

summary_table2 = io_operations.groupby(['data_size_category', 'access_pattern']).agg({
    'latency_ms': ['count', 'mean', 'median', 'min', 'max', 'std']
}).round(4)

print(summary_table2)

# Summary table 3: Combined analysis
print("\n" + "=" * 80)
print("SUMMARY TABLE 3: Average Latency by Operation Type, Data Size, and Access Pattern")
print("=" * 80)

summary_table3 = io_operations.groupby(['operation_type', 'data_size_category', 'access_pattern']).agg({
    'latency_ms': ['count', 'mean', 'median', 'std']
}).round(4)

print(summary_table3)
```
